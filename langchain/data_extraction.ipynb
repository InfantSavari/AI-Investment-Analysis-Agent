```python
import yfinance as yf
import os, json
from mcp.server.fastmcp import FastMCP
import requests
import logging
import pandas as pd
```

```python
mcp = FastMCP("AI Investing Agent Server")
logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s [%(levelname)s] %(message)s",  
    handlers=[
        logging.FileHandler("app.log"),  
        logging.StreamHandler()         
    ]
)
dataApi = os.environ.get("ALPHA_VANTAGE_API")
```

```python
class DataTools:
    def __init__(self):
        logging.info("Tools class initialised")
        
    def get_income_stmt(self, ticker: str):
        """ this Function retrieves income statement through Alpha Vantage api """
        url = f"https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol={ticker}&apikey={dataApi}"
        
        rawData = requests.get(url) # requesting data through rest api call
        jsonData = rawData.json() # response from alpha vantage is converted into json

        quarterlyData = jsonData["annualReports"][:3] # only last 3 yrs are considered

        """ Conversion of json data into markdown data format. first into dataframe and setting column for  """
        markdowned_Data = pd.DataFrame(quarterlyData) 
        markdowned_Data.set_index("FiscalDateEnding", inplace = True)
        markdowned_Data = markdowned_Data.to_markdown()
        print(markdowned_Data)
        logging.info("annual income statements are retrieved")

    def get_balance_sheet(self, ticker: str):
        url = f"https://www.alphavantage.co/query?function=BALANCE_SHEET&symbol={ticker}&apikey={dataApi}"

        rawData = requests.get(url)
        jsonData = rawData.json()

        quarterlyData = jsonData["annualReports"][:3]

        markdowned_Data = pd.DataFrame(quarterlyData)
        markdowned_Data.set_index("FiscalDateEnding", inplace = True)
        markdowned_Data = markdowned_Data.to_markdown()
        print(markdowned_Data)
        logging.info("annual Balance statements are retrieved")

    def get_cashflow_stmt(self, ticker: str):
        url = f"https://www.alphavantage.co/query?function=CASH_FLOW&symbol={ticker}&apikey={dataApi}"

        rawData = requests.get(url)
        jsonData = rawData.json()

        quarterlyData = jsonData["annualReports"][:3]

        markdowned_Data = pd.DataFrame(quarterlyData)
        markdowned_Data.set_index("FiscalDateEnding", inplace = True)
        markdowned_Data = markdowned_Data.to_markdown()
        print(markdowned_Data)
        logging.info("annual cashflow statements are retrieved")
        
                
```
